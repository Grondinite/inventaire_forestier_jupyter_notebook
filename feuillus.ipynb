### Synthèse de la biomasse des feuillus

Créer :
1. Un tableau global avec moyenne, médiane et somme de la colonne `biomass`.
2. Un tableau avec les mêmes indicateurs, calculés séparément pour chaque espèce (`species`).

```python
biomass_summary = df_feuillus['biomass'].agg(['mean', 'median', 'sum']).to_frame(name='biomass')
biomass_summary
```

```python
biomass_by_species = df_feuillus.groupby('species', as_index=True)['biomass'].agg(['mean', 'median', 'sum'])
biomass_by_species
```
